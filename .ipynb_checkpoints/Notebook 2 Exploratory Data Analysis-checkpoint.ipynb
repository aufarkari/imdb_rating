# Importing Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

# Reading Data

```python
data = pd.read_csv('C:/Users/karteuku/Desktop/IMDB_rating/imdb_rating/data/data.csv')

data.set_index('tconst',inplace= True)
```

```python
data.shape
```

# Exploratory Data Analysis

## Title Type Analysis

### TV Episode is the most common title type (47%)

```python
data['titleType'].value_counts()/(data['titleType'].notna().sum())*100
```

### TV Episode has the highest average rating

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x = 'titleType', y = 'averageRating', data = data)
```

## Genre Analysis

```python
genre_col_list = [col for col in data.columns if col.startswith('genre_')]

len(genre_col_list)
```

```python
diff_in_mean_dict = {}
p_values_dict = {}

for col in genre_col_list:
    mask_0 = data[col] ==0
    mask_1 = data[col] ==1
    
    diff_in_mean = data['averageRating'][mask_1].mean() - data['averageRating'][mask_0].mean()
    diff_in_mean_dict[col] = diff_in_mean
    
    p_values = stats.ttest_ind(data['averageRating'][mask_1], data['averageRating'][mask_0])
    p_values_dict[col] = p_values

diff_in_mean_df = pd.DataFrame.from_dict(diff_in_mean_dict,orient='index')
diff_in_mean_df.columns =['diff_in_mean']
diff_in_mean_df['p_values'] = [values[1] for values in p_values_dict.values()]
```

### Documentary, History, and Crime are the genres with most positive difference in average rating

```python
diff_in_mean_df.sort_values(by='diff_in_mean', ascending = False).head(5)
```

### Horror, Thriller, and Adult are the genres with most negative difference in average rating

```python
diff_in_mean_df.sort_values(by='diff_in_mean', ascending = True).head(5)
```

## Correlation Analysis

```python
## Creating list to store numerical column and non-numerical column 
object_col_list = []
num_col_list = []

for col in data.columns:
    if data[col].dtypes =='object':
        object_col_list.append(col)
    else :
        num_col_list.append(col)
```

```python
## Creating list to store continuous column and dummy column 
dummy_col_list = []
continuous_col_list = []

for col in num_col_list:
    if len(data[col].unique()) > 2:
        continuous_col_list.append(col)
    else :
        dummy_col_list.append(col)
```

```python
len(dummy_col_list)
```

### Number of unique language, unique ordering, unique region, unique titles, and unique types are highly correlated

```python
sns.heatmap(data[continuous_col_list].corr())
```

```python
sns.heatmap(data[continuous_col_list].corr().abs())
```

### Average show past rating and median past rating of the director shows the highest correlation with average rating

```python
pd.DataFrame(data[continuous_col_list].corr()['averageRating'].sort_values(ascending=False)).head(5)
```

```python
pd.DataFrame(data[continuous_col_list].corr()['averageRating'].sort_values(ascending=False)).tail(5)
```

```python
sns.heatmap(pd.DataFrame(data[continuous_col_list].corr()['averageRating'].sort_values(ascending=False)))
```
